<h1> < style="color:#FF0000";> GIT </h1>
Aplicacion para versionamiento de fuentes que nos permite llevar el control  de los cambios que se realizan a los  proyectos de desarrollo de software, una ventaja de este tipo de aplicativos es que nos permite tener un control de los cambios  que se han realizado durante  la vida del proyecto, como quien modifico cierta funcion  y cuando la modifico,  hacer un rollback   hacia una version anterior y a demas de ello y una de las mas importartes es tener en un unico repositorio de la ultima version de los fuentes mas actualizados, suele sucede que cuando no exite un lugar centralizado  donde los fuentes no estan actualizado y los mismo se utilizan para una nueva implementacion  es muy  posible que cambios que ya se habian hecho sean borrados, o bien cuando los nuevos cambios estan en proceso de validacion se comprueba que no tiene la funcionalidad requerida debido a fallas ya que no tomaron en cuenta  las nuevas validaciones o fucniones ya implementadas, esto comumento suece sucede aquellas organizaciones que cuentan con aplicativos bastante robusto y no se lleva un contro de fuente o bien si se lleva  los mismo no estan estan siendo actualizado al repositorio central.
